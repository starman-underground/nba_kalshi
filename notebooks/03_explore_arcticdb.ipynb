```python
import os
import polars as pl
import dtale
from arcticdb import Arctic
from pprint import pprint

# Initialize Arctic
local_store = Arctic("lmdb://../data/")

# Get all libraries in the store
libraries = local_store.list_libraries()
print(libraries)
```

```python
# Drop test_lib if it exists
if "test_lib" in libraries:
    local_store.delete_library("test_lib")

# Show the libraries
libraries = local_store.list_libraries()
print(libraries)
```

```python
# Show the symbols in the library
symbols = local_store["raw_market_metadata"].list_symbols()
pprint(f"Found {len(symbols)} symbols in the library")
for symbol in symbols:
    print(f"Symbol: {symbol}, length: {len(local_store['raw_market_metadata'].read(symbol).data)}")
```

```python
# open market_metadata in dtale
raw_market_metadata = local_store["raw_market_metadata"]
raw_game_market_metadata = raw_market_metadata.read("KXNBAGAME", output_format="POLARS")

# dtale expects pandas; convert at the display boundary
dtale.show(raw_game_market_metadata.data.to_pandas())
```
